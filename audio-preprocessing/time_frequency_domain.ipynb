```python
import librosa
import matplotlib.pyplot as plt
```

```python
from ipynb.fs.full.preparation import get_librosa_audio, get_audio_path

audio_paths = get_audio_path()
audio_librosa = get_librosa_audio()
```

```python
def calculate_spectrogram(wave):
    X = librosa.stft(wave)
    Xdb = librosa.amplitude_to_db(abs(X))
    return Xdb

def display_spectrogram(emotion, waves):
    fig, ax = plt.subplots(nrows = len(waves), ncols = 1, sharex = True)
    fig.suptitle(emotion)
    for i, wave in enumerate(waves):
        ax[i].set_title(audio_paths[emotion][i])
        Xdb = calculate_spectrogram(wave[0])
        librosa.display.specshow(Xdb, sr = wave[1], x_axis = 'time', y_axis = 'hz',ax = ax[i])
    fig.tight_layout()

for key, values in audio_librosa.items():
    display_spectrogram(key, values)
```
