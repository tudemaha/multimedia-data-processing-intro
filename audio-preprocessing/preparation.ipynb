```python
import librosa
```

```python
audio_list = {
    'happy': ['0102', '0104', '0107'],
    'neutral': ['0103', '0107', '0110'],
    'sad': ['0104', '0106', '0109']
}
```

```python
def get_audio_path():
    audio_paths = {}
    for key, values in audio_list.items():
        audio_paths[key] = [f'SpeechEmotion/{key}/{key}-{value}.wav' for value in values]
    return audio_paths
```

```python
from IPython.display import Audio

audio_paths = get_audio_path()

for key, values in audio_paths.items():
    print(key.upper())
    for value in values:
        display(Audio(value))
```

```python
def get_librosa_audio():
    audio_librosa = {}
    for key, values in audio_paths.items():
        audio_librosa[key] = []
        for value in values:
            y, sr = librosa.load(value)
            audio_librosa[key].append((y, sr))

    return audio_librosa
```
